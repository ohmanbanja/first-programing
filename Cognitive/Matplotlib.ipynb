```python
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))
plt.plot(x, np.sin(x-5), color='chartreuse')
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x,x+0,'-g')
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x,y, c=color, s=sizes, alpha=0.3)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
```

```python
plt.scatter(features[0],features[1], alpha=0.2,s=features[3]*100, cmap='viridis',
           c=iris.target)

plt.xlabel(iris.features_names[0])
plt.ylabel(iris.features_names[1])

```

```python
#errorbar
x = np.linspace(0,10,50)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x,y, yerr=dy,c='k' )
```

```python
plt.style.use('seaborn-whitegrid')
plt.errorbar(x, y, yerr=dy, fmt='o' , c='y',
            ecolor='blue', elinewidth=3)
```
